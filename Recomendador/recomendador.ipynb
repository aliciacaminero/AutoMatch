```python
df_recomendador = dataset_coches.copy()
```

```python
# Número de filas antes de eliminar duplicados
print(f"Número de filas antes de eliminar duplicados: {df_recomendador.shape[0]}")

# Eliminar duplicados
duplicados = df_recomendador.duplicated().sum()
print(f"Número de filas duplicadas: {duplicados}")
if duplicados > 0:
    df_recomendador = df_recomendador.drop_duplicates()

# Número de filas después de eliminar duplicados
print(f"Número de filas después de eliminar duplicados: {df_recomendador.shape[0]}")
```

```python
 # 1. Mantener solo las columnas necesarias
columns_to_keep = [
        'make', 'model', 'version', 'fuel', 'year', 'kms', 
        'power', 'shift', 'price', 
        'dealer_name', 'dealer_address', 'dealer_city', 'dealer_zip_code', 'province'
    ]
df_recomendador = df_recomendador[columns_to_keep]
print(df_recomendador.head())
```

```python
# 2. Tratamiento de valores nulos

# # Para combustible, usamos el valor más común

df_recomendador['fuel'] = df_recomendador['fuel'].fillna(df_recomendador['fuel'].mode()[0])
print(df_recomendador)
```

```python
# Para kilometraje, usamos la mediana por año y modelo
df_recomendador['kms'] = df_recomendador.groupby(['make', 'model', 'year'])['kms'].transform(
    lambda x: x.fillna(x.median()) if not x.median() != x.median() else x.fillna(df_recomendador['kms'].median())
)
print(df_recomendador)
```

```python
# Para potencia, usamos la mediana por modelo y versión
df_recomendador['power'] = df_recomendador.groupby(['make', 'model', 'version'])['power'].transform(
        lambda x: x.fillna(x.median()) if not x.median() != x.median() else x.fillna(df_recomendador['power'].median())
)
print(df_recomendador)  
```

```python
# 3. Calcular la antigüedad del coche
current_year = datetime.now().year
df_recomendador['age'] = current_year - df_recomendador['year']
print(df_recomendador)
```

```python
# 2. Normalizar con los decimales especificados

# Para kilometraje (2 decimales)
df_recomendador['kms'] = df_recomendador['kms'].round(2)
    
# Para potencia (1 decimal)
df_recomendador['power'] = df_recomendador['power'].round(1)
    
# Para precio (2 decimales)
df_recomendador['price'] = df_recomendador['price'].round(2)

print(df_recomendador)
```

```python
# 6. Asegurarse de que los tipos de datos son correctos
df_recomendador['dealer_zip_code'] = df_recomendador['dealer_zip_code'].fillna(0).astype(int)

# Para código postal (0 decimales)
df_recomendador['dealer_zip_code'] = df_recomendador['dealer_zip_code'].round(0)
print(df_recomendador)
```

```python
# 5. Crear columnas derivadas útiles para el recomendador
df_recomendador['price_per_power'] = np.ceil(df_recomendador['price'] / df_recomendador['power'] * 100) / 100
current_year = datetime.now().year
df_recomendador['price_per_year'] = np.ceil(df_recomendador['price'] / (current_year - df_recomendador['year']) * 100) / 100

# 2. Recalculamos los ratios
df_recomendador['price_per_power'] = np.ceil((df_recomendador['price'] / df_recomendador['power']) * 100) / 100
df_recomendador['price_per_year'] = np.ceil((df_recomendador['price'] / (current_year - df_recomendador['year'])) * 100) / 100

# 3. Normalizamos los ratios en las mismas columnas
df_recomendador['price_per_power'] = ((df_recomendador['price_per_power'] - df_recomendador['price_per_power'].min()) / (df_recomendador['price_per_power'].max() - df_recomendador['price_per_power'].min())).round(2)
    
df_recomendador['price_per_year'] = ((df_recomendador['price_per_year'] - df_recomendador['price_per_year'].min()) / (df_recomendador['price_per_year'].max() - df_recomendador['price_per_year'].min())).round(2)

print(df_recomendador)
```

```python
# 7. Crear una columna con la información completa del dealer para Streamlit
df_recomendador['dealer_info'] = df_recomendador.apply(
        lambda x: f"{x['dealer_name']}\n{x['dealer_address']}\n{x['dealer_city']}, {x['province']}\nCP: {x['dealer_zip_code']}", 
        axis=1
)
print(df_recomendador)
```

```python
# Primero manejamos los nulos en power
power_median = df_recomendador.groupby(['make', 'model'])['power'].transform('median')
df_recomendador['power'] = df_recomendador['power'].fillna(power_median)
df_recomendador['power'] = df_recomendador['power'].fillna(df_recomendador['power'].median())  # Para grupos sin mediana
print(df_recomendador)
```

```python
print(df_recomendador.isnull().sum())
```

```python
def handle_final_missing_values(df_recomendador):
    """
    Maneja los últimos valores nulos del dataset con verificaciones detalladas.
    """
    # 1. Manejar información del dealer usando código postal
    print("\n1. Procesando información del dealer...")

    # Crear diccionario de mapeo
    zip_to_location = df_recomendador[df_recomendador['dealer_city'].notna()].groupby('dealer_zip_code').agg({
        'dealer_city': 'first',
        'province': 'first'
    }).to_dict('index')

    print(f"Número de códigos postales únicos mapeados: {len(zip_to_location)}")

    # Rellenar ciudad y provincia
    nulos_ciudad_antes = df_recomendador['dealer_city'].isnull().sum()

    for idx in df_recomendador.index:
        if pd.isna(df_recomendador.loc[idx, 'dealer_city']):
            zip_code = df_recomendador.loc[idx, 'dealer_zip_code']
            if zip_code in zip_to_location:
                df_recomendador.loc[idx, 'dealer_city'] = zip_to_location[zip_code]['dealer_city']
                df_recomendador.loc[idx, 'province'] = zip_to_location[zip_code]['province']

    nulos_ciudad_despues = df_recomendador['dealer_city'].isnull().sum()
    print(f"Nulos en dealer_city antes: {nulos_ciudad_antes}, después: {nulos_ciudad_despues}")

    # 2. Manejar shift (transmisión)
    print("\n2. Procesando shift...")
    nulos_shift_antes = df_recomendador['shift'].isnull().sum()

    # Inferir por marca/modelo/año
    df_recomendador['shift'] = df_recomendador.groupby(['make', 'model', 'year'])['shift'].transform(
        lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else 'Unknown')
    )

    nulos_shift_intermedio = df_recomendador['shift'].isnull().sum()

    # Usar "Unknown" para los restantes
    df_recomendador['shift'] = df_recomendador['shift'].fillna('Unknown')

    nulos_shift_despues = df_recomendador['shift'].isnull().sum()
    print(f"Nulos en shift antes: {nulos_shift_antes}, después de inferencia: {nulos_shift_intermedio}, final: {nulos_shift_despues}")

    # 3. Manejar version
    print("\n3. Procesando version...")
    nulos_version_antes = df_recomendador['version'].isnull().sum()

    # Crear versión estándar
    df_recomendador['version'] = df_recomendador['version'].fillna(
        df_recomendador['make'] + ' ' + df_recomendador['model'] + ' Standard'
    )

    nulos_version_despues = df_recomendador['version'].isnull().sum()
    print(f"Nulos en version antes: {nulos_version_antes}, después: {nulos_version_despues}")

    # 4. Manejar dealer_name, dealer_address, dealer_city, province
    print("\n4. Procesando dealer_name, dealer_address, dealer_city, province...")
    df_recomendador['dealer_name'] = df_recomendador['dealer_name'].fillna('Unknown Dealer')
    df_recomendador['dealer_address'] = df_recomendador['dealer_address'].fillna('Unknown Address')
    df_recomendador['dealer_city'] = df_recomendador['dealer_city'].fillna('Unknown City')
    df_recomendador['province'] = df_recomendador['province'].fillna('Unknown Province')

    nulos_dealer_name_despues = df_recomendador['dealer_name'].isnull().sum()
    nulos_dealer_address_despues = df_recomendador['dealer_address'].isnull().sum()
    nulos_dealer_city_despues = df_recomendador['dealer_city'].isnull().sum()
    nulos_province_despues = df_recomendador['province'].isnull().sum()
    print(f"Nulos en dealer_name después: {nulos_dealer_name_despues}")
    print(f"Nulos en dealer_address después: {nulos_dealer_address_despues}")
    print(f"Nulos en dealer_city después: {nulos_dealer_city_despues}")
    print(f"Nulos en province después: {nulos_province_despues}")

    # 5. Manejar price_per_power
    print("\n5. Procesando price_per_power...")
    df_recomendador['price_per_power'] = df_recomendador.apply(
        lambda row: math.ceil(row['price'] / row['power'] * 100) / 100 if pd.notnull(row['price']) and pd.notnull(row['power']) and row['power'] != 0 else 0, axis=1
    )

    nulos_price_per_power_despues = df_recomendador['price_per_power'].isnull().sum()
    print(f"Nulos en price_per_power después: {nulos_price_per_power_despues}")

    # 6. Verificar y ajustar price_per_year si es necesario
    print("\n6. Verificando price_per_year...")
    df_recomendador['price_per_year'] = df_recomendador.apply(
        lambda row: math.ceil(row['price'] / (2025 - row['year']) * 100) / 100 if pd.notnull(row['price']) and (2025 - row['year']) != 0 else 0, axis=1
    )
    
    nulos_price_per_year_despues = df_recomendador['price_per_year'].isnull().sum()
    print(f"Nulos en price_per_year después: {nulos_price_per_year_despues}")

    # 7. Conversión de tipos de datos
    print("\n7. Convertir tipos de datos...")
    df_recomendador['dealer_zip_code'] = df_recomendador['dealer_zip_code'].astype(str)
    cols_categoricas = ['make', 'model', 'version', 'fuel', 'shift', 'dealer_name', 'dealer_address', 'dealer_city', 'province', 'dealer_info']
    for col in cols_categoricas:
        df_recomendador[col] = df_recomendador[col].astype('category')

    # Verificación final
    print("\nEstado final del dataset:")
    print("\nValores nulos finales:")
    print(df_recomendador.isnull().sum())

    print("\nDistribución de valores en shift:")
    print(df_recomendador['shift'].value_counts(normalize=True).round(3))

    return df_recomendador

# Uso de la función
print("Iniciando procesamiento...")
df_recomendador = handle_final_missing_values(df_recomendador)
print("\nProcesamiento completado.")
```
